# XGBoost

XGBoost (Extreme Gradient Boosting) is an efficient and popular machine learning algorithm known for its speed and performance in both regression and classification tasks. It belongs to the family of boosting algorithms and has gained widespread adoption in various machine learning competitions and real-world applications due to its effectiveness.

Key Features of XGBoost:
1. Gradient Boosting Framework: XGBoost follows a gradient boosting framework, which sequentially combines multiple weak learners (usually decision trees) to create a strong predictive model.

2. Regularization: It includes L1 (Lasso) and L2 (Ridge) regularization terms in its objective function to control model complexity and prevent overfitting.

3. Tree Pruning: Implements tree pruning strategies to control tree growth and minimize overfitting. It allows early stopping to find the optimal number of boosting rounds.

4. Handling Missing Values: It can handle missing data within the dataset during training.

5. Parallel Processing: Offers efficient parallel and distributed computing capabilities, making it faster compared to traditional gradient boosting algorithms.

Advantages of XGBoost:

1. Highly Accurate: Often yields superior performance compared to other algorithms.
2. Speed and Scalability: Efficient implementation leads to fast training and prediction times.
3. Robustness: Handles outliers and missing values well.
4. Feature Importance: Provides insights into feature importance, aiding in feature selection.

Parameters to Tune:

1. n_estimators: Number of boosting rounds or weak learners.
2. learning_rate: Controls the contribution of each weak learner.
3. max_depth: Maximum depth of each tree in the ensemble.
4. subsample, colsample_bytree: Controls subsampling of data and features to prevent overfitting.
5. reg_alpha, reg_lambda: Regularization parameters.

Applications:

1. Finance: Stock market prediction, credit risk modeling.
2. Marketing: Customer churn prediction, recommendation systems.
3. Healthcare: Disease diagnosis, drug discovery.

XGBoost's ability to handle large datasets, its predictive power, and efficient optimization make it a go-to choice for various machine learning tasks. However, parameter tuning and understanding the trade-offs between model complexity and performance are crucial for maximizing its potential.

### XGBoost Classifier 

```python
import numpy as np
from sklearn.datasets import make_classification
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
```

```python
# Generating synthetic data for classification
X, y = make_classification(n_samples=100, n_features=2, n_classes=2, 
                           n_clusters_per_class=1, n_redundant=0, random_state=42)
```

```python
X
```

```python
y
```

```python
# Creating an XGBoost Classifier
clf = XGBClassifier(n_estimators=100, max_depth=3, random_state=42)
clf.fit(X, y)

# Predictions
y_pred = clf.predict(X)

# Calculating and printing accuracy score
accuracy = accuracy_score(y, y_pred)
print(f'Accuracy of XGBoost Classifier is : {accuracy:.4f}')
```

```python
# Plotting decision boundaries 
plt.figure(figsize=(8, 6))
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1), np.arange(y_min, y_max, 0.1))
Z = clf.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)
plt.contourf(xx, yy, Z, alpha=0.3)
plt.scatter(X[:, 0], X[:, 1], c=y, edgecolor='black')
plt.title('XGBoost Classifier')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

### XGBoost Regressor

```python
import numpy as np
import matplotlib.pyplot as plt
from xgboost import XGBRegressor
from sklearn.metrics import r2_score
```

```python
# Generating synthetic data for regression
np.random.seed(0)
X = np.sort(5 * np.random.rand(80, 1), axis=0)
y = np.sin(X).ravel() + np.random.randn(80) * 0.1
```

```python
X
```

```python
y
```

```python
# Creating an XGBoost Regressor
reg = XGBRegressor(n_estimators=100, max_depth=3, random_state=42)
reg.fit(X, y)

# Predictions
y_pred = reg.predict(X)

# Checking Accuracy
r2_score = r2_score(y,y_pred)
print(f'The r2_score value of XGBoost Regressor Model is {r2_score:.2f}')
```

```python
# Plotting the results
plt.figure(figsize=(8, 6))
plt.scatter(X, y, s=20, edgecolor="black", c="darkorange", label="data")
plt.plot(X, y_pred, color="cornflowerblue", label="prediction")
plt.xlabel("data")
plt.ylabel("target")
plt.title("XGBoost Regressor")
plt.legend()
plt.show()
```
